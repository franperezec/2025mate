```python
import pandas as pd
import numpy as np
from IPython.display import display


# Variables
W1, W2 = 100, 200
GK1, GK2 = 300, 400
P1, P2 = 5, 10
C1H1, C1H2 = 3, 2
C2H1, C2H2 = 1, 4


import sympy as sp
sp.init_printing(use_unicode=True, use_latex='mathjax', order='lex')


# Define symbols
w, r, q1, q2 = sp.symbols('w r q1 q2')
a1, b1, a2, b2 = sp.symbols('a1 b1 a2 b2')
K, L = sp.symbols('K L')
def pro_1(K, L, a1, b1):
    result =  K**b1 * L**a1
    return result

def pro_2(K, L, a2, b2):
    result =  K**b2 * L**a2
    return result




display(pro_1(K,L,a1,b2))

# a1, b1, a2, b2 = 0.4, 0.6, 0.3, 0.7
K, L, lambda1, lambda2 = sp.symbols('K L lambda1 lambda2')

# Equations based on the production function
equation1 = sp.Eq(w, lambda1 * a1 * L**(a1 - 1) * K**b1)
equation2 = sp.Eq(r, lambda1 * b1 * K**(b1 - 1) * L**a1)
equation3 = sp.Eq(q1, K**b1 * L**a1)

equation4 = sp.Eq(w, lambda2 * a2 * L**(a2 - 1) * K**b2)
equation5 = sp.Eq(r, lambda2 * b2 * K**(b2 - 1) * L**a2)
equation6 = sp.Eq(q2, K**b2 * L**a2)

# Solve the system of equations
#display("primera eq")
solution1 = sp.solve([equation1, equation2, equation3], (K, L, lambda1), dict=True)
#display(solution1)

solution2 = sp.solve([equation4, equation5, equation6], (K, L, lambda2), dict=True)

display("Solution for System 1:")
display(solution1)
display("Solution for System 2:")
display(solution2)
```
